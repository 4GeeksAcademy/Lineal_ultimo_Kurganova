Linear Regression model

```python
train_data = pd.read_csv("../data/processed/clean_trainMM.csv")
test_data = pd.read_csv("../data/processed/clean_testMM.csv")

train_data.head()
```

```python
X_train_M = train_data.drop(["charges"], axis = 1)
y_train = train_data["charges"]
X_test_M = test_data.drop(["charges"], axis = 1)
y_test = test_data["charges"]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_M, y_train)
```

```python
print(f"Intercep (a): {model.intercept_}")
print(f"Coefficients (b1, b2): {model.coef_}")
```

```python
y_pred = model.predict(X_test_M)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(f"MSE: {mean_squared_error(y_test, y_pred)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

Cuanto menor sea el valor del RMSE, mejor será el modelo. Un modelo perfecto (un modelo hipotético que siempre pueda predecir el valor esperado exacto) tendría un valor para esta métrica de 0.
